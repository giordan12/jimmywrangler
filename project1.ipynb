### Summary of what I did in this project
#### My original goal was to find out if the world cup games had any effect on the online shopping numbers in Brazil. The last world cup player there was in 2014. Therefore, I used data of all the world cup games and their results. I isolated the games for the 2014 world cup and focused on the games in which Brazil played. However, there were only about 4 games that Brazil played and this seemed too little of a sample to perform any analysis. Therefore I focused on analyzing the online shopping database that I found. It was not complicated to clean the data and display it in a nice way. However, I did notice a weird outlier which made no sense in the data. For the date '2017-11-24' there was an extremely large amount of shopping. This did not follow any important world cup games so I did not understand why that value was so large. However, after performing some research online I discovered that Brazil also celebrates black friday. I still was not sure but after watching videos of people fighting for stuff on supermarkets I convinced myself that this is true.
#### Unfortunately I did not find any effect that the soccer games had in the shopping habits of Brazilian people but I did learn something new about Brazil.
